## REMOVING OUTLIERS 

```python
plt.figure(figsize=(16,5))
plt.subplot(121)
sns.distplot(data['length_of_service'])
plt.subplot(122)
sns.distplot(data['age'])
plt.show()
```

```python
percentile25_los=data['length_of_service'].quantile(0.25)
percentile75_los=data['length_of_service'].quantile(0.75)
percentile25_age=data['age'].quantile(0.25)
percentile75_age=data['age'].quantile(0.75)
print("Q2 of los:",percentile25_los)
print("Q4 of los:",percentile75_los)
print("Q2 of age:",percentile25_age)
print("Q4 of age:",percentile75_age)
```

```python
iqr_los=percentile75_los-percentile25_los
print("IQR of los",iqr_los)
iqr_age=percentile75_age-percentile25_age
print("IQR of age",iqr_age)
```

```python
upper_limit_los=percentile75_los+1.5*iqr_los
lower_limit_los=percentile25_los-1.5*iqr_los
upper_limit_age=percentile75_age+1.5*iqr_age
lower_limit_age=percentile25_age-1.5*iqr_age
```

```python
print("upper limit of los",upper_limit_los)
print("lower limit of los",lower_limit_los)
print("upper limit of age",upper_limit_age)
print("lower limit of age",lower_limit_age)
```

```python
data[data['length_of_service']>13]
```

```python
data[(data['age']>54) | (data['age']<14)]
```

### CAPPING OF OUTLIERS

```python
data['length_of_service']=np.where(
data['length_of_service']>upper_limit_los,
upper_limit_los,
np.where(
data['length_of_service']<lower_limit_los,
lower_limit_los,data['length_of_service'])
)
```

```python
data['length_of_service'].describe()
```

```python
sns.boxplot(data['length_of_service'])
plt.show()
```

```python
data['age']=np.where(
data['age']>upper_limit_age,upper_limit_age
    ,np.where(
    data['age']<lower_limit_age,lower_limit_age,
    data['age']))
```

```python
sns.boxplot(data['age'])
plt.show()
```
